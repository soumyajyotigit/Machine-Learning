```python
train_df = pd.read_csv('/kaggle/input/titanic/train.csv')
test_df=pd.read_csv('/kaggle/input/titanic/test.csv')
```

```python
#save passenger id 
passengerId=test_df['PassengerId'].values
```

```python
# Drop unnecessary cols
train_df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
test_df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
train_df.isnull().sum()
```

```python
test_df.isnull().sum()
```

```python
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].mean())
test_df['Age'] = test_df['Age'].fillna(test_df['Age'].mean())
```

```python
train_df['Embarked'] = train_df['Embarked'].fillna('S')
test_df['Fare'] = test_df['Fare'].fillna(test_df['Fare'].mean())
```

```python
train_df.isnull().sum()
```

```python
test_df.isnull().sum()
```

```python
train_df['family']=train_df['Parch']+train_df['SibSp'] +1
test_df['family']=test_df['Parch']+test_df['SibSp']+1
```

```python
def family_size(number):
    if number == 1:
        return "Alone"
    elif number>1 and number<=4:
        return "Small"
    else:
        return "Large"
```

```python
train_df['family_size'] = train_df['family'].apply(family_size)
test_df['family_size'] = test_df['family'].apply(family_size)
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
train_df.drop(columns=['Parch','SibSp','family'],inplace=True)
```

```python
test_df.drop(columns=['Parch','SibSp','family'],inplace=True)
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
train_df = pd.get_dummies(train_df,columns=['Pclass','Sex','Embarked','family_size'],drop_first=True)
test_df = pd.get_dummies(test_df,columns=['Pclass','Sex','Embarked','family_size'],drop_first=True)
```

```python
# Train dataset
X = train_df.iloc[:,1:].values
y = train_df.iloc[:,0].values
```

```python
# Test dataset
X_final = test_df.iloc[:,:].values
```

```python
X_final.shape
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(max_iter=500,C=100)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
y_final = clf.predict(X_final)
```
